```python
import joblib
import pandas as pd
import numpy as np

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
new_data = pd.read_csv('./Data/Cubic_unseen.csv')
new_data
```

```python
model = joblib.load('./CubicModel.pkl')
print(model)
```

```python
pred = model.predict(new_data)
pred
new_data['pred']= pd.DataFrame(pred)
new_data
```

```python
r2 = r2_score(new_data.price, pred)
mae = mean_absolute_error(new_data.price, pred)
mean_act = new_data.price.mean()
mean_pred = pred.mean()
min_act = min(new_data.price)
max_act = max(new_data.price)
mape = mean_absolute_percentage_error(new_data.price, pred)
print(f'R2: {r2}\nmae: {mae}\nact_mean: {mean_act}\npred_mean: {mean_pred}\nmape: {mape}\nact_range: {min_act} to {max_act}')
```
